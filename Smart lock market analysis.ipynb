```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
path = "D:\smart_loccksasmart_lockss.csv"
data = pd.read_csv(path)
data
```

```python
data['prices'] = data['prices'].apply(lambda x: f"₹{x}")
```

```python
data.head(10)
```

```python
data['sku'] = data['product_name'].str.split().str.get(0)
data['sku']
```

```python
data['ratings'] = data['ratings'].str.split().str.get(0)
data['ratings']
```

```python
number_of_brands = data['sku'].nunique()
print(f"Number of brands: {number_of_brands}")
```

```python
sorted_data = data.sort_values(by=['ratings', 'count_of_ratings'], ascending=[False, False])

# Display the sorted DataFrame
print(sorted_data)
```

```python
sku_count_per_product_name = data.groupby('product_name')['sku'].count()
print(sku_count_per_product_name )
```

```python
data['ratings'] = pd.to_numeric(data['ratings'], errors='coerce')
average_ratings = data.groupby('sku')['ratings'].mean().reset_index().round(1)
average_ratings
```

```python
price_bands = [
    ('Less than INR 4999', 0, 4999),
    ('INR 3000-4999', 3000, 4999),
    ('INR 5000-9999', 5000, 9999),
    ('INR 10000-14999', 10000, 14999),
    ('INR 15000-19999', 15000, 19999),
    ('Greater than 20000', 20000, float('inf'))
]

# Convert 'prices' column to numeric format
data['prices'] = data['prices'].replace('[\₹,]', '', regex=True).astype(float)

# Categorize SKUs into price bands
sku_counts = []
for band in price_bands:
    band_name, lower_bound, upper_bound = band
    count = data[(data['prices'] >= lower_bound) & (data['prices'] <= upper_bound)].shape[0]
    sku_counts.append((band_name, count))

# Display the count of SKUs in each price band
for band_name, count in sku_counts:
    print(f"{band_name}: {count}")
```
